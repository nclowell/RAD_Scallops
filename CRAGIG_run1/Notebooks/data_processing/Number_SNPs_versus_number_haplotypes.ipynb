## Relationship between number of SNPs and haplotypes

**20170428**

One would expect, that if SNPs within tags exhibit no linkage, then the number of haplotypes would increase exponentially with the number of SNPs. If there were total linkage, then the number of haplotypes would remain the same (2) despite an increase in SNPs. Here, I took a look at that relationship in the first lane of scallop data.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cd /mnt/hgfs/SHARED_FOLDER/WorkingFolder
```

```python
# define functions for plotting, no linkage and total linkage
def tot_link(x):
    y = x*x
    return y
def no_link(x):
    y = 2 - .5/x
    return y
```

```python
hapfile = open("Stacks_2/batch_100.haplotypes.tsv", "r")
lines = hapfile.readlines()
lines = lines[1:] # exclude the header line
hapfile.close()
```

```python
# initiate dictionary
mydict = {}

cat_IDs = []

for line in lines:
    linelist = line.split() # split on white space
    key = linelist[0] # get catalog ID as key for each key value pair in dictionary
    cat_IDs.append(key)
    hapcols = linelist[2:] # get a list of just the lines of columns that have haplotype information
    col_list = [] # iniatiate a list that will include all haplotypes in a line - later, I will use set() for just the unique haplotypes, and then use that to get a count of unique haplotypes
    for col in hapcols: # for each column within a row of the haplotype file, count SNPs and alleles
        cell_list = col.split("/") # for non-homozygotes
        for item in cell_list:
            if item != '-': #  and item != 'consensus'
                col_list.append(item)
    unique_haps_in_locus = list(set(col_list))
    num_unique_haps = len(unique_haps_in_locus)
    lengths = []
    for un_hap in unique_haps_in_locus:
        length = len(un_hap)
        lengths.append(length)
    if len(list(set(lengths))) > 1:
        print list(set(lengths))
        print unique_haps_in_locus
    mydict[key] = [unique_haps_in_locus, num_unique_haps, lengths[0]] # each item in dictionary is set of unique haplotypes, number of unique haplotypes, number of SNPs in locus

snp_counts = []
hap_counts = []

for locus in cat_IDs:
    if mydict[locus][0] != ['consensus']:
        numhap = mydict[locus][1]
        hap_counts.append(numhap)
        numsnp = mydict[locus][2]
        snp_counts.append(numsnp)
```

```python
# plot results
plt.scatter(snp_counts,hap_counts, alpha=.25) # multiple box plots with list of lists
plt.xlabel('Number of SNPs in a RAD tag') # axis label
plt.ylabel('Number of haplotypes in a RAD tag') # axis label
plt.suptitle('Number of SNPs by number of haplotypes across individuals in RAD tags') # title

# define axes limits
ax=plt.axes()
ax.set_xlim([0,50])
ax.set_ylim([0,50])

# add curves, no and total linkage
x = np.linspace(0.1,50,1000)
ax.plot(x, tot_link(x),color = 'g')
ax.plot(x, no_link(x), color = 'm')

# add legend
plt.legend(["No linkage","Total linkage","Empirical data"], loc= (.6,.1))

plt.show() # show plot
```
